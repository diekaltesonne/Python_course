```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CompleteDataset.csv')
df.head()
```

```python
bins=pd.cut(df['RB'],np.arange(0,150,2))
```

```python
mv=df.groupby(bins)['RB'].agg(['count','mean'])
mv_2=df.groupby(bins)['ST'].agg(['count','mean'])
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Average RB and ST value')
ax1.plot(mv['count'],mv['mean'],c='blue')
ax1.set(xlabel='Amount of players', ylabel='Mean RB of the player in a bin')
ax2.plot(mv_2['count'],mv_2['mean'],c='blue')
ax2.set(xlabel='Amount of players', ylabel='Mean CT of the player in a bin')
```
