```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('CompleteDataset.csv')
df.columns
df.head()
```

```python
agg_1=df.groupby(['Club'])['CB'].agg(['count','mean','sum'])
agg_2=df.groupby(['Club'])['Overall'].agg(['count','mean','sum'])
```

```python
graph=plt.scatter(x=agg_1['mean'], y=agg_2['mean'], s=agg_1['sum']/100)
plt.title('Correlation between average Overall and average CB grouped by clubs')
plt.ylabel('Mean CB')
plt.xlabel('Players in a club')
```
